## Task 1: Conceptual Understanding (Mandatory – Written)


### 1. "Love" vs "love"

- Case-sensitive difference 
- Models treat them as different tokens unless normalized 
- "Love" → start of sentence / proper noun vs "love" → general word

### 2. If stopwords are not removed

- Adds noise 
- Increases dimensionality 
- Reduces model performance (especially in TF-IDF)

### 3. When removing stopwords is harmful

- Sentiment analysis → "not good" becomes "good" 
- Chatbots / QA → meaning lost ("what is your name")

### 4. Stemming vs Lemmatization

- Stemming → crude cut (running → runn)
- Lemmatization → dictionary-based (running → run)
- Lemmatization is more accurate but slower

## Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = re.sub(r"http\S+|www\S+|@\S+", "", text)

    text = re.sub(r"\d+", "", text)

    
    text = text.lower() # Lowercase

    text = re.sub(r"(.)\1{2,}", r"\1", text)
    
    text = re.sub(r"[^a-z\s]", "", text) # Remove special characters (keep words)

    text = re.sub(r"\s+", " ", text).strip() # Remove extra spaces

    tokens = text.split()

    tokens = [w for w in tokens if len(w) > 2 or w in ["no", "not", "i"]] # Remove short tokens (<=2 except 'no', 'not','I')

    clean_sentence = " ".join(tokens)

    return tokens,clean_sentence
```

```python
print(preprocess_text("I have 2 dogs"))
print(preprocess_text("This    is    good"))
print(preprocess_text("soooo goooood!!!"))
print(preprocess_text("WOW!!! This is GREAT!!!"))
print(preprocess_text("Visit http://example.com now"))
print(preprocess_text("Email me at test@gmail.com"))
print(preprocess_text("Nooooo I am not happy"))
print(preprocess_text("OK OK OK I got it"))
print(preprocess_text("Win $$$ now!!! Limited offer!!!"))
print(preprocess_text("Call me at 9876543210"))
```

## Task 3: Stress Testing

```python
# Test Sentences (covers all required cases)
test_sentences = [
    "Get 100% FREE access now!!!",              # numbers + symbols
    "I absolutely looooved this product 😍😍",  # emojis + repeated chars
    "Worst service ever... 0/10",               # numbers + punctuation
    "Call me at 9876543210",                   # numbers
    "This is THE best course!!!",              # mixed case
    "Visit https://openai.com now!",           # URL
    "Nooooo this is baaad!!!",                 # repeated chars
    "OK OK OK I got it",                       # short tokens
    "Win $$$ now!!! Limited offer!!!",         # symbols
    "I am not happy with this"                 # stopword edge case
]

# Stress Testing Output
for idx, text in enumerate(test_sentences, 1):
    
    result = preprocess_text(text)   # avoid unpack error
    
    tokens = result[0]
    clean_sentence = result[1]

    print(f"\nSentence {idx}")
    print("Original Text   :", text)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("-" * 50)
```

## Task 4: Token Analytics


```python
import numpy as np

analytics = []

for idx, text in enumerate(test_sentences, 1):
    
    result = preprocess_text(text)  
    
    tokens = result[0]
    clean = result[1]

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(t) for t in tokens]) if tokens else 0

    analytics.append({
        "Sentence No": idx,
        "Original": text,
        "Total Tokens": total_tokens,
        "Unique Tokens": unique_tokens,
        "Avg Token Length": round(avg_length, 2)
    })

for a in analytics:
    print(a)
```

## Task 5: Frequency Analysis


```python
from collections import Counter

all_tokens = []

for text in test_sentences:
    result = preprocess_text(text)   
    tokens = result[0]
    all_tokens.extend(tokens)
counter = Counter(all_tokens)

top_10 = counter.most_common(10)

least_5 = counter.most_common()[-5:]

print("Top 10 Most Frequent Words:")
print(top_10)

print("\nTop 5 Least Frequent Words:")
print(least_5)
```

## Task 6: Build Full Pipeline


```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)   

        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(test_sentences)

print(output["tokens"])
print(output["clean_sentences"])
```
